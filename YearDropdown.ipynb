```python
import pandas as pd
import datetime as dt
```

```python
# Pull in fireball csv data
fireball_years = pd.read_csv("Resources/Fireball_Data_Years.csv")
fireball_years
```

```python
#pull in meterorite data
fireball_years = fireball_years.rename(columns ={'Peak Brightness Date/Time (UT)':'year'})
fireball_years
```

```python
#drop columns to year, for bullet altitude graph to graph mean, max, min for each year
fireball_years=fireball_years.drop(columns=['Unnamed: 0','Latitude (deg.)','Longitude (deg.)','Velocity (km/s)','vx','vy','vz','Total Radiated Energy (J)','Calculated Total Impact Energy (kt)'])
```

```python
fireball_years
```

```python
fb_count = fireball_years.groupby(['year',], as_index=False)['Altitude (km)'].nunique()
fb_count
```

```python
#pull out year and altitude
fireball_years=fireball_years.dropna()
fireball_years
```

```python
#group by year
altitude_mean = fireball_years.groupby(['year',], as_index=False)['Altitude (km)'].mean()
altitude_mean = altitude_mean.rename(columns ={'Altitude (km)':'Alt_mean_km'})
altitude_mean.head(5)
```

```python
altitude_min = fireball_years.groupby(['year',], as_index=False)['Altitude (km)'].min()
altitude_min = altitude_min.rename(columns ={'Altitude (km)':'Alt_min_km'})
altitude_min.head(5)
```

```python
altitude_max = fireball_years.groupby(['year',], as_index=False)['Altitude (km)'].max()
altitude_max = altitude_max.rename(columns ={'Altitude (km)':'Alt_max_km'})
altitude_max.head(5)
```

```python
#merge min and max
min_mean=altitude_min.merge(altitude_mean, how='outer').round(1)
min_mean.head(5)
```

```python
#merge in mean
min_mean_max=min_mean.merge(altitude_max, how='outer').round(1)
min_mean_max.head(5)
```

```python
#merge in count
fb_altitude=min_mean_max.merge(fb_count, how='outer').round(1)
fb_altitude = fb_altitude.rename(columns ={'Altitude (km)':'fb_count'})
fb_altitude.head(5)
```

```python
fb_altitude =fb_altitude.sort_values(['year'], ascending=False)
fb_altitude.head(5)
```

```python
fb_altitude.to_csv('Resources/fb_altitude.csv')
```

```python
#meterorite for year db only
m = pd.read_csv("Resources/Meteorite_Landings_Years.csv")
m
```

```python
#fireball for year db only
fb = pd.read_csv("Resources/Fireball_Data_Years.csv")
fb = fb.rename(columns ={'Peak Brightness Date/Time (UT)':'year'})
fb
```

```python
#merge fireballs and meteorite for year df
merge=fb.merge(m, how='outer')
merge
```

```python
#drop extra columns
merge = merge.drop(columns={'Unnamed: 0','Latitude (deg.)','Longitude (deg.)','Altitude (km)','Velocity (km/s)','vx','vy','vz','Total Radiated Energy (J)','Calculated Total Impact Energy (kt)','name','id','nametype','recclass','mass (g)','fall','reclat','reclong','GeoLocation'})
merge
```

```python
#place in descending order
year =merge.sort_values(['year'], ascending=False)
year
```

```python
#drop duplicates
year=year.drop_duplicates()
year
```

```python
#drop year 2101 or index place [0]
year=year.drop(year.index[0])
year
```

```python
#save to csv
year.to_csv('Resources/year.csv')
```
